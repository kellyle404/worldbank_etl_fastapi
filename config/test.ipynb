```python
from sqlalchemy import create_engine
import pandas as pd
from config.config import DATABASE_URL

engine = create_engine(DATABASE_URL)

query = "SELECT t.id AS topic_id, t.name AS topic_name, im.code AS indicator_code, im.name AS indicator_name, c.name AS country_name, iv.date, iv.value FROM topics t JOIN indicator_meta im ON t.id = im.topic_id JOIN indicator_values iv ON im.id = iv.indicator_id JOIN countries c ON c.id = iv.country_id ORDER BY random() LIMIT 10;"

df = pd.read_sql(query, engine)
display(df)
```
